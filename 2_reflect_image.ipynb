```python
from matplotlib import image
from matplotlib import pyplot

# load image as pixel array
im = image.imread('a.jpg')

print("type(im)           ", type(im))
print("im.shape           ", im.shape)
print("type(im[1][1][1])  ", type(im[1][1][1]))

pyplot.imshow(im)
pyplot.show()
```

```python
import numpy

# Reflect the image (each row has an even number of pixels)
narr = numpy.zeros((len(im), len(im[0]), 3), dtype=numpy.uint8)
for i in range (0, len(narr)):
    for j in range (0, len(narr[i])):
        for k in range(0, len(im[i][j])):
            narr[i][j][k] = im[i][-j][k]
            narr[i][-j][k] = im[i][j][k]

pyplot.imshow(narr)
pyplot.show()
```

```python
from PIL import Image

# create Pillow image from numpy array
image2 = Image.fromarray(narr)

print("type(image2)          ", type(image2))
print("image2                ", image2)
print("image2.format         ", image2.format)
print("image2.mode           ", image2.mode)
print("image2.size           ", image2.size)
print("image2.getdata        ", image2.getdata)
print("image2.getdata()      ", image2.getdata())
print("image2.getdata()[20]  ", image2.getdata()[20])
print(" ")

image2 = Image.fromarray(narr).save('a_reflect.png')
```
